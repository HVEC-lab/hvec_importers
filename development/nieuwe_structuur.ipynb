```python
import requests
import pandas as pd

URL = "https://ddapi20-waterwebservices.rijkswaterstaat.nl/METADATASERVICES/OphalenCatalogus"

payload = {
    "CatalogusFilter": {
        "Compartimenten": True,
        "Grootheden": True,
        "Eenheden": True,
        "Hoedanigheden": True,
        "Parameters": True,
        "ProcesTypes": True
    }
}

response = requests.post(URL, json=payload, timeout=60)
response.raise_for_status()

data = response.json()

locaties = data.get("LocatieLijst", [])
df_locaties = pd.DataFrame(locaties)

print(df_locaties.head())
print(f"Aantal meetlocaties: {len(df_locaties)}")
print(df_locaties.columns.tolist())
```
